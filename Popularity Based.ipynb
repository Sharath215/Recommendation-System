```python
import pandas as pd
import numpy as np
```

```python
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
Movies = pd.read_csv('ml-100k/u.item',sep='|',names = i_cols, encoding ='latin-1')
```

```python
Movies
```

```python
Movies = Movies.drop(['release date','video release date', 'IMDb URL','unknown'], axis=1)
```

```python
Movies.shape
```

```python
Movies.head(5)
```

```python
#read the ratings dataset
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
Ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=r_cols,encoding='latin-1')
```

```python
Ratings
```

```python
Ratings = Ratings.drop(['unix_timestamp','user_id'],axis = 1)
```

```python
Ratings.head(10)
```

```python
movie_Category=['movie id','movie title','Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
```

```python
movie_Category
```

```python
Ratings['rating'].value_counts()
```

```python
# movies ID with above 4 ratings

Movie_ID_Popular = Ratings[Ratings['rating']>4]['movie_id']
```

```python
Movie_ID_Popular = np.array(Movie_ID_Popular)
```

```python
Movie_ID_Popular # got movies with rating >4
```

```python
len(Movie_ID_Popular)
```

```python
Unique_Movie_ID_Popular = list(set(Movie_ID_Popular))
```

```python
len(Unique_Movie_ID_Popular)
```

```python
Unique_Movie_ID_Popular
```

```python
Unique_Movie_ID_Popular = np.array(Unique_Movie_ID_Popular)
```

```python
Unique_Movie_ID_Popular
```

```python
Movies_Names = pd.DataFrame(columns = Movies.columns)
```

```python
Movies_Names # emoty dataframe
```

```python
Unique_Movie_ID_Popular = pd.DataFrame(Unique_Movie_ID_Popular, columns=['ID'])
```

```python
# copies the needed movies from the original table based on the popular movie id value

Movies_Names = pd.merge(Movies, Unique_Movie_ID_Popular, left_on='movie id', right_on='ID', how='inner')
```

```python
Movies_Names # movies which users gave top rating (>=4)
```
